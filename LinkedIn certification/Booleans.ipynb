```python
weatherIsNice = True
haveUmbrella = False

if not (haveUmbrella or weatherIsNice):
    print('Stay inside')
else: 
    print('Go for a walk')
```

```python
weatherIsNice = True
haveUmbrella = False

if not haveUmbrella and not weatherIsNice:
    print('Stay inside')
else: 
    print('Go for a walk')
```
